```python
import seaborn as sns
```

```python
# Set the style and color palette
sns.set_style("whitegrid")
sns.set_palette("pastel")
```

```python
# Distribution of Sales
plt.figure(figsize=(8,6))
sns.histplot(df['SALES'], kde=True, color='blue',palette='pastel')
plt.title('Distribution of Sales')
plt.xlabel('Sales')
plt.ylabel('Frequency')
plt.show()
```

```python
#Quantity Ordered by Deal Size: This will show the relationship between QUANTITYORDERED and DEALSIZE.
plt.figure(figsize=(8, 6))
sns.lineplot(x='DEALSIZE', y='QUANTITYORDERED', data=df,palette='pastel')
plt.title('Quantity Ordered by Deal Size')
plt.xlabel('Deal Size')
plt.ylabel('Quantity Ordered')
plt.show()
```

```python
#Sales Over Time (Month-wise): We can see how sales have varied over the months.
plt.figure(figsize=(8, 6))
sns.lineplot(x='MONTH_ID', y='SALES', data=df)
plt.title('Sales Over Months')
plt.xlabel('Month')
plt.ylabel('Sales')
plt.show()
```

```python
#Sales by Product Line: Compare sales across different product lines.
plt.figure(figsize=(8, 6))
sns.barplot(x='PRODUCTLINE', y='SALES', color= 'violet', data=df, estimator=sum)
plt.title('Total Sales by Product Line')
plt.xticks(rotation=45)
plt.xlabel('Product Line')
plt.ylabel('Total Sales')
plt.show()
```

```python
#Deal Size Count Plot: Count the occurrences of each DEALSIZE.
plt.figure(figsize=(6, 6))
sns.countplot(x='DEALSIZE', data=df,palette='pastel')
plt.title('Count of Deal Size')
plt.xlabel('Deal Size')
plt.ylabel('Count')
plt.show()
```

```python
#Sales by Country: This can show which countries generate the most sales
plt.figure(figsize=(8, 6))
sns.barplot(x='COUNTRY', y='SALES', data=df, estimator=sum)
plt.title('Total Sales by Country')
plt.xticks(rotation=90)
plt.xlabel('Country')
plt.ylabel('Total Sales')
plt.show()
```

```python
#Sales vs. MSRP: Investigate the relationship between the MSRP (Manufacturer's Suggested Retail Price) and SALES.
plt.figure(figsize=(8, 6))
sns.scatterplot(x='MSRP', y='SALES', data=df,palette='pastel')
plt.title('Sales vs. MSRP')
plt.xlabel('MSRP')
plt.ylabel('Sales')
plt.show()
```

```python
#Sales by Year: Total sales broken down by year
plt.figure(figsize=(6, 6))
sns.barplot(x='YEAR_ID', y='SALES', data=df, estimator=sum,palette='pastel')
plt.title('Total Sales by Year')
plt.xlabel('Year')
plt.ylabel('Total Sales')
plt.show()
```
